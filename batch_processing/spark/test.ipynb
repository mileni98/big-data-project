```python
#https://stackoverflow.com/questions/849211/shortest-distance-between-a-point-and-a-line-segment
#https://www.youtube.com/watch?v=egmZJU-1zPU

def dist(x1, y1, x2, y2, x3, y3): # x3,y3 is the point
    px = x2-x1
    py = y2-y1

    norm = px*px + py*py

    u =  ((x3 - x1) * px + (y3 - y1) * py) / float(norm)

    if u > 1:
        u = 1
    elif u < 0:
        u = 0

    x = x1 + u * px
    y = y1 + u * py

    dx = x - x3
    dy = y - y3

    # Note: If the actual distance does not matter,
    # if you only want to compare what this function
    # returns to other results of this function, you
    # can just return the squared distance instead
    # (i.e. remove the sqrt) to gain a little performance

    dist = (dx*dx + dy*dy)**.5

    return dist
```

```python
import matplotlib.pyplot as plt
def plot_segmented_line(x1, y1, x2, y2, x3, y3):
    plt.plot([x1, x2], [y1, y2], marker='o', label='Segmented Line')
    plt.plot(x3, y3, marker='o', color='red', label='Point (x3, y3)')
    
    closest_x, closest_y = x1, y1
    closest_distance = dist(x1, y1, x2, y2, x3, y3)
    for i in range(100):
        t = i / 100.0
        current_x = x1 + t * (x2 - x1)
        current_y = y1 + t * (y2 - y1)
        current_distance = dist(x1, y1, x2, y2, current_x, current_y)
        
        if current_distance < closest_distance:
            closest_x, closest_y = current_x, current_y
            closest_distance = current_distance

    plt.plot([x3, closest_x], [y3, closest_y], linestyle='--', color='green', label='Closest Distance')
    
    plt.legend()
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Visualization of dist Function')
    plt.grid(True)
    plt.show()
```

```python
x = dist(39.659,138.085, 39.832,138.352, 39, 139)
print(x)
plot_segmented_line(39.659,138.085, 39.832,138.352, 39, 139)
```
